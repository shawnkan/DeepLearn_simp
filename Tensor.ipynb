```python
import tensorflow as tf
import numpy as np
```

```python
#step -1 zero rank tensor
tensor1=tf.constant(4)
print(tensor1)
```

```python
# rank one tensor
tensor_rank1=tf.constant([2.0,9.0,7.0])
tensor_rank1

#1D array
```

```python
# create a matrix/ rank2 tensor
tensor_rank2=tf.constant([[0,1,2],[2,3,4],[4,5,6]])
tensor_rank2
```

```python
# convert to numpy array
tensor_rank2.numpy()
```

```python
a = tf.constant([[1,2],[3,4]])
b= tf.constant([[5,6],[7,8]])

print(tf.add(a,b))
```

```python
print(tf.multiply(a,b))
```

```python
#maximum value of the tensor
b=tf.constant([[5,6],[7,8]])
print(tf.reduce_max(b))
```

```python
print(tf.math.argmax(b))
```

```python
# broadcasting in tensors
x=tf.constant([1,2,3])
y=tf.constant(2)
print(tf.multiply(x,2))
print(tf.multiply(x,y))
```

```python
# reshape the tensor
b=tf.constant([[5,6,6],[7,8,9]])
print(b.shape)
```

```python
new_tensor=tf.reshape(b,[3,2])
new_tensor
```

```python
new_tensor.ndim
```

```python
tensor4.ragged_rank
```
