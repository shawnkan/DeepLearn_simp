```python
# load the dataset
from tensorflow.keras.datasets import mnist
print('Data Loaded')
```

```python
# create train & test
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
# normalize the pixel in 0 to 1
X_train=X_train/255.
X_test=X_test/255.
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Dense,Flatten
```

```python
# readibility better
# multi in and multi out better to use sequential

def build_model_function():
  input_layer=Input(shape=(28,28))
  flatten_layer=Flatten()(input_layer)
  first_dense=Dense(128,activation='relu')(flatten_layer)
  output_layer=Dense(10,activation='softmax')(first_dense)
  func_model=Model(inputs=input_layer,outputs=output_layer)
  return func_model
```

```python
model = build_model_function()
model.summary()
```

```python
# compile the model
# use sparse instead because never one hot.. multi class classification
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print('model is compiled')
```

```python
# train the model
batch_size=128
epochs=10
```

```python
history=model.fit(X_train,y_train,
                  batch_size=batch_size,
                  epochs=epochs,
                  validation_data=(X_test,y_test))
```

```python
score_test=model.evaluate(X_test,y_test)
print('test score',score_test)
```
