```python
# load the dataset
from tensorflow.keras.datasets import mnist
print('Data Loaded')
```

```python
# create train & test
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0],cmap='gray')
plt.show()
```

```python
# Preprocessing
# Reshape the training & test data
X_train_reshaped = X_train.reshape(60000,784)
X_test_reshaped = X_test.reshape(10000,784)
```

```python
X_train_reshaped.shape
```

```python
# normalize the pixel in 0 to 1
X_train_reshaped=X_train_reshaped/255.
X_test_reshaped=X_test_reshaped/255.
```

```python
y_train[:5]
```

```python
# convert class vectors into one hot encoding
from tensorflow.keras.utils import to_categorical
num_classes=10
y_train_encoded=to_categorical(y_train,num_classes)
y_test_encoded=to_categorical(y_test,num_classes)
```

```python
y_train_encoded[:5]
```

```python
# 1. import required lib
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense #ann layers are called dense in tf
print('imported')
```

```python
# Create Architecture - 3 layers, 100, 64 ,32 , activation relu , output layer-softmax
model = Sequential()
#first hidden layer
model.add(Dense(100, activation='relu',input_shape=(784,)))
#second hidden layer
model.add(Dense(64, activation='relu'))
#third hidden layer
model.add(Dense(32,activation='relu'))
#fourth hidden layer_ see if accuracy changes
model.add(Dense(32,activation='relu'))

#output layer
model.add(Dense(10,activation='softmax'))
```

```python
# (input shape * weights) + total bias
```

```python
model.summary()
```

```python
for layer in model.layers:
  if hasattr(layer,'weights'):
    weights,biases=layer.get_weights()
    print('shape',weights.shape)
    print(f'weights of {layer.name}:')
    print(weights)
    print('shape',biases.shape)
    print(f'bias of {layer.name}:')
    print(biases)
```

```python
# compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
print('model is compiled')
```

```python
#train the model
batch_size=128
epochs=5
```

```python
#60000/128 so 468.75 about 1 epoch 469 iterations
# every iteration the weights are changing
```

```python
history = model.fit(X_train_reshaped, y_train_encoded,
                    batch_size=batch_size,
                    epochs=epochs,
                    validation_data=(X_test_reshaped, y_test_encoded))
```

```python
score_test=model.evaluate(X_test_reshaped,y_test_encoded)
print('test score', score_test)
```

```python
score_train=model.evaluate(X_train_reshaped,y_train_encoded)
print('train score',score_train)
```

```python
for layer in model.layers:
  if hasattr(layer,'weights'):
    weights,biases=layer.get_weights()
    print('shape',weights.shape)
    print(f'weights of {layer.name}:')
    print(weights)
    print('shape',biases.shape)
    print(f'bias of {layer.name}:')
    print(biases)
```

```python
# plot the graph
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train','test'],loc='best')
plt.show()
```

```python
# plot the graph
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model losses')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend(['train','test'],loc='best')
plt.show()
```

```python
# predictions
y_pred=model.predict(X_test_reshaped)
```

## Early Stopping and Dropouts

```python
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Create Architecture - 3 layers, 100, 64 ,32 , activation relu , output layer-softmax
model1 = Sequential()
#first hidden layer
model1.add(Dense(100, activation='relu',input_shape=(784,)))
model1.add(Dropout(0.3))
#second hidden layer
model1.add(Dense(64, activation='relu'))
model1.add(Dropout(0.2))

#third hidden layer
model1.add(Dense(32,activation='relu'))
model1.add(Dropout(0.25))
#fourth hidden layer_ see if accuracy changes
model1.add(Dense(32,activation='relu'))
model1.add(Dropout(0.25))

#output layer
model1.add(Dense(10,activation='softmax'))
```

```python
model1.summary()
```

```python
# compile the model
model1.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print('model is compiled')
```

```python
early_stop=EarlyStopping(monitor='val_loss',
                         min_delta=0.01,
                         mode='min',
                         verbose=1,
                         patience=10
                         )
```

```python
epochs = 50
```

```python
history=model1.fit(X_train_reshaped,y_train_encoded,
                  batch_size=batch_size,
                  epochs=epochs,
                  validation_data=(X_test_reshaped,y_test_encoded),
                  callbacks=[early_stop]
                  )
```

```python
score_train=model1.evaluate(X_train_reshaped,y_train_encoded)
print('train score',score_train)
score_test=model1.evaluate(X_test_reshaped,y_test_encoded)
print('test score',score_test)
```

```python
# plot the graph
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train','test'],loc='best')
plt.show()
```

```python
# plot the graph
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model losses')
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.legend(['train','test'],loc='best')
plt.show()
```
