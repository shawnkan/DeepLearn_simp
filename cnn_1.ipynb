```python
#load the data
from tensorflow.keras.datasets import mnist
(X_train, y_train), (X_test, y_test)=mnist.load_data()
print('Data Loaded')
```

```python
X_train.shape
```

```python
#reshape the data, one colour
X_train = X_train.reshape(60000,28,28,1)
X_train = X_train.reshape(60000,28,28,1)
```

```python
#convert y into one hot labels
num_classes=10
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

```python
y_train[:5]
```

```python
#create CNN Architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
print('Imported')
```

```python
#create a 2-layer CNN
model = Sequential()

#first CNN layer
model.add(Conv2D(64,kernel_size=(3,3), activation='relu', input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))

#second CNN layer
model.add(Conv2D(32,kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#Flatten
model.add(Flatten())
#create ANN
model.add(Dense(1000,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
#compile the model
model.compile(loss='categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
print('compiled')
```

```python
model.summary()
```

```python
#input shape-(kernel-1)
# 28-(3-1) so 26,26

# (Shape of filter + bias)*total number of filters
# ((3*3)+1)*64 =640 param
#1 bias

#13-(3-1) so 11,11

#64 previous outputs...
#((3*3*64)+1)*32
```

```python
#run the model
history = model.fit(X_train, y_train, epochs=3, batch_size=128, validation_data=(X_test, y_test))
```

```python
#evaluate the model
model.evaluate(X_test,y_test)
```
